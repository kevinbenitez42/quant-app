```python
import requests

import matplotlib.pyplot as plt

def get_crypto_historical_prices(coin):
    url = f"https://api.coingecko.com/api/v3/coins/{coin}/market_chart"
    params = {
        "vs_currency": "usd",
        "days": 30
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        crypto_prices = data["prices"]
        return crypto_prices
    else:
        print(f"Failed to fetch {coin} historical prices.")

cardano_historical_prices = get_crypto_historical_prices("cardano")
ethereum_historical_prices = get_crypto_historical_prices("ethereum")

if cardano_historical_prices:
    print("Cardano Historical Prices:")
    cardano_timestamps = [price[0] for price in cardano_historical_prices]
    cardano_prices = [price[1] for price in cardano_historical_prices]

    # Calculate cumulative returns for Cardano
    cardano_cumulative_returns = [((price - cardano_prices[0]) / cardano_prices[0]) * 100 for price in cardano_prices]

    # Plotting the cumulative returns for Cardano
    plt.plot(cardano_timestamps, cardano_cumulative_returns)
    plt.xlabel('Timestamp')
    plt.ylabel('Cumulative Returns (%)')
    plt.title('Cardano Cumulative Returns')
    plt.show()

if ethereum_historical_prices:
    print("Ethereum Historical Prices:")
    ethereum_timestamps = [price[0] for price in ethereum_historical_prices]
    ethereum_prices = [price[1] for price in ethereum_historical_prices]

    # Calculate cumulative returns for Ethereum
    ethereum_cumulative_returns = [((price - ethereum_prices[0]) / ethereum_prices[0]) * 100 for price in ethereum_prices]

    # Plotting the cumulative returns for Ethereum
    plt.plot(ethereum_timestamps, ethereum_cumulative_returns)
    plt.xlabel('Timestamp')
    plt.ylabel('Cumulative Returns (%)')
    plt.title('Ethereum Cumulative Returns')
    plt.show()
    import plotly.graph_objects as go

    bitcoin_ytd_returns = ((prices[-1] - prices[0]) / prices[0]) * 100
    ethereum_ytd_returns = ((ethereum_prices[-1] - ethereum_prices[0]) / ethereum_prices[0]) * 100
    cardano_ytd_returns = ((cardano_prices[-1] - cardano_prices[0]) / cardano_prices[0]) * 100

    fig = go.Figure()

    fig.add_trace(go.Scatter(x=[timestamps[0], timestamps[-1]], y=[0, bitcoin_ytd_returns], mode='lines', name='Bitcoin'))
    fig.add_trace(go.Scatter(x=[ethereum_timestamps[0], ethereum_timestamps[-1]], y=[0, ethereum_ytd_returns], mode='lines', name='Ethereum'))
    fig.add_trace(go.Scatter(x=[cardano_timestamps[0], cardano_timestamps[-1]], y=[0, cardano_ytd_returns], mode='lines', name='Cardano'))

    fig.update_layout(title='Year-to-Date Returns',
                      xaxis_title='Timestamp',
                      yaxis_title='YTD Returns (%)')

    fig.show()
```
