```python
# https://www.quantrocket.com/codeload/quant-finance-lectures/quant_finance_lectures/Lecture42-Introduction-to-Pairs-Trading.ipynb.html
import numpy as np
import pandas as pd

import statsmodels
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint

import matplotlib.pyplot as plt
import yfinance as yf
from Quantapp.Algorithms import Algorithms
algorithms = Algorithms()
plt.rcParams["figure.figsize"] = (20, 7)

ticker1 = yf.Ticker('ual').history(period='3y')['Close']
ticker2 = yf.Ticker('uso').history(period='3y')['Close']
```

```python
pd.concat([ticker1,ticker2],axis=1).plot();
pd.concat([ticker1.pct_change(50),ticker2.pct_change(50)],axis=1).plot()
plt.axhline(y=0, color='r', linestyle='--');
pd.concat([ticker1.pct_change(200),ticker2.pct_change(200)],axis=1).plot()
plt.axhline(y=0, color='r', linestyle='--');
```

```python
spread = (ticker2.pct_change(50) - ticker1.pct_change(50))

spread.plot() # Plot the spread
#plt.axhline(spread.mean(), color='red', linestyle='--') # Add the mean
plt.xlabel('Time')

spread = (ticker2.pct_change(200) - ticker1.pct_change(200))
spread.plot() # Plot the spread
#plt.axhline(spread.mean(), color='red', linestyle='--') # Add the mean
plt.xlabel('Time')
plt.legend(['Price Spread: 50D','Price Spread: 200D'])
plt.axhline(y=0, color='r', linestyle='--');
```

```python
rolling_correlation_21  = algorithms.rolling(algorithms.correlation, 21, ticker1, ticker2)
rolling_correlation_50  = algorithms.rolling(algorithms.correlation, 50, ticker1, ticker2)
rolling_correlation_200 = algorithms.rolling(algorithms.correlation, 200, ticker1, ticker2)

rolling_correlation_21.plot()
rolling_correlation_50.plot()
rolling_correlation_200.plot()

plt.legend(['21D Corr','50D Corr', '200D Corr'])
plt.axhline(y=0, color='r', linestyle='-')
```

```python

rolling_cointegration_50  = algorithms.rolling(algorithms.cointegration, 50, ticker1, ticker2)
rolling_cointegration_200 = algorithms.rolling(algorithms.cointegration, 200, ticker1, ticker2)

rolling_cointegration_50.plot();
rolling_cointegration_200.plot();
plt.axhline(y=0.1, color='r', linestyle='-');

plt.legend(['50D Coint', '200D Coint'])
```
